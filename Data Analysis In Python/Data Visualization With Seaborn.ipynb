### Data Visualization With Seaborn

Seaborn is apython visualization library on Matplotlib that provides a high-leve; interface for drawing attractive and informative statistical graphics. Seaborn helps in creating complex visualizations with just few lines of code.

```python
import seaborn as sns

```

```python
excercise = sns.load_dataset('exercise')
excercise
```

```python
import matplotlib.pyplot as plt
sns.scatterplot(x='id',y='pulse',data = excercise)
plt.title("Scatter Plot by Seaborn")
```

```python
sns.barplot(x='diet',y='time',data=excercise)
plt.title("Bar Plot of Diet and Time")
plt.show()
```

```python
sns.boxplot(x='diet',y='time',data=excercise)
plt.title("Box Plot of Diet and Time")
plt.show()
```

```python
sns.violinplot(x='diet',y='time',data=excercise)
plt.title("Violinplot Plot of Diet and Time")
plt.show()
```

```python
sns.histplot(excercise['time'],bins = 10,kde=True)
plt.title("Histplot Plot of Diet and Time")
plt.show()
```

```python
sns.pairplot(excercise)
```
