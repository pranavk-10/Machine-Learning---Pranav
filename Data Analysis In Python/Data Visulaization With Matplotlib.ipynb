````python
import matplotlib.pyplot as plt
````

````python
x = [1,2,3,4,5]
y = [1,4,7,9,16]

plt.plot(x,y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("Matplotlib Graph")
plt.show()
````

````python
x = [1,2,3,4,5]
y = [1,4,7,9,16]

plt.plot(x,y,color = 'red',linestyle = '--')
plt.show()
````

````python
x = [1,2,3,4,5]
y = [1,4,7,9,16]

plt.plot(x,y,color = 'red',linestyle = '--',marker = 'o')
plt.show()
````

````python
x = [1,2,3,4,5]
y = [1,4,7,9,16]

plt.plot(x,y,color = 'red',linestyle = '--',marker = 'o')
plt.grid(True)
plt.show()
````

````python
x = [1 , 2 , 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [2,4,6,8,10]

plt.figure(figsize=(9,5)) # figure size with respect to pixels
plt.subplot(1,2,1)
plt.plot(x,y1,color = 'green')
plt.title("PLOT X AND Y1")

plt.subplot(1,2,2)
plt.plot(x,y2,color = 'green')
plt.title("PLOT X AND Y1")

plt.show()
````

Chalo step-by-step breakdown karte hain is code ka:

### 1. **Libraries Import (Missing in your code)**  
Sabse pehle, plotting ke liye `matplotlib.pyplot` library import karte hain:

```python
import matplotlib.pyplot as plt
```

### 2. **Data Setup:**  
- `x = [1, 2, 3, 4, 5]`: Ye x-axis ke values hain.
- `y1 = [1, 4, 9, 16, 25]`: Ye pehle plot ke y-axis values hain (perfect squares).
- `y2 = [2, 4, 6, 8, 10]`: Ye doosre plot ke y-axis values hain (even numbers).

### 3. **Figure Size Set Karna:**  
```python
plt.figure(figsize=(9,5))
```
- `figsize=(9,5)` ka matlab hai ki plot ka size 9 units wide aur 5 units tall hoga. Ye inches mein hota hai, pixels nahi.

### 4. **Subplot Banana:**  
```python
plt.subplot(1,2,1)
```
- `plt.subplot(1, 2, 1)` ka matlab:
  - **1** row,
  - **2** columns,
  - Aur hum **first** subplot bana rahe hain.
  
  Iska matlab hai ki tumhare plot ka layout kuch aise dikhega:

  ```
  [ Plot 1 | Plot 2 ]
  ```

### 5. **Plotting the Data:**  
```python
plt.plot(x, y, color='green')
```
- Yaha tum `x` aur `y` plot kar rahe ho, lekin tumhare code me `y` define nahi hai.  
  **Error aayega.** Shayad tumhe `y1` use karna tha.

  Corrected code:
  ```python
  plt.plot(x, y1, color='green')
  ```

- `color='green'`: Plot green color mein dikhaya jayega.

### 6. **Plot Title Set Karna:**  
```python
plt.title("PLOT X AND Y1")
```
- Ye tumhare plot ka title hai.

### 7. **Plot Display Karna:**  
```python
plt.show()
```
- Ye command final plot screen pe dikhata hai.

### **Corrected and Complete Code:**

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]

y2 = [2, 4, 6, 8, 10]

plt.figure(figsize=(9,5))  # Set figure size

plt.subplot(1, 2, 1)       # First subplot
plt.plot(x, y1, color='green')  # Corrected to y1
plt.title("PLOT X AND Y1")

plt.show()
```

````python
categories = ['a','b','c','d','e']
values = [1,2,3,4,5]

plt.bar(categories,values,color = 'black',linewidth = 3)
plt.show()
````

### Histogram

Histogram are used to represent the distribution of a dataset . They divide the data into bins and count the number of data points in each bins

````python
x = [1,2,3,4,5,6,7,8,5,4,3,2,5,5,7]
plt.hist(x,bins = 5,color = 'orange',edgecolor = 'black')
#plt.show()
````

````python
x =[1,2,3,4,5]
y = [2,3,4,5,6]

plt.scatter(x,y,color = 'teal',marker='*')
````

````python
labels = ['Vishwajeet','Aziz','Aman','Aamish']
values = [0.25,0.6,1.3,1.0]
explode = (0.1,0,0,0)

plt.pie(values,explode = explode,labels=labels)
plt.show()
````
