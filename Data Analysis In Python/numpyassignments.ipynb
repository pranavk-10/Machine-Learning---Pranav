````python
import numpy as np

array = np.random.randint(1,21,size = (5,5))
print("Original Array")
print(array)

array[:,2]=1
print("Modified Array")
print(array)

````

This code snippet is modifying a NumPy array by replacing all elements in the third column with `1`. Let's break it down step by step:

### Explanation:

1. **`array[:, 2] = 1`**  
   - `array` is assumed to be a NumPy array.
   - `:` in `array[:, 2]` means **"select all rows"**.
   - `2` in `array[:, 2]` refers to the **third column** (indexing starts from 0).
   - `= 1` sets all values in this column to `1`.

2. **`print("Modified array:")`**  
   - Displays the message `"Modified array:"` before printing the updated array.

3. **`print(array)`**  
   - Prints the modified array to the console.

### Example:

#### Before Modification:
```python
import numpy as np

array = np.array([[5, 8, 3], 
                  [2, 6, 9], 
                  [4, 7, 5]])

print("Original array:")
print(array)
```
**Output:**
```
Original array:
[[5 8 3]
 [2 6 9]
 [4 7 5]]
```

#### After Modification:
```python
array[:, 2] = 1  # Replacing third column with 1
print("Modified array:")
print(array)
```
**Output:**
```
Modified array:
[[5 8 1]
 [2 6 1]
 [4 7 1]]
```

### Key Takeaways:
- The slicing operation `array[:, 2]` selects the entire third column.
- Assigning `1` to `array[:, 2]` updates all elements in that column.
- Useful for modifying specific columns in multi-dimensional arrays.

### **Understanding `array[1:-1, 0]`**  
👉 Yeh **"middle rows ke first column"** ko select karega.  

#### **Step-by-Step Breakdown:**
- **`1:-1` → Row selection:**  
  - Yeh **first (0th) aur last (-1th) row ko exclude karega.**
  - Sirf **1st row se lekar (n-2)th row tak ka data milega.**
  
- **`, 0` → Column selection:**  
  - Yeh **first column (index `0`) ko select karega.**

#### **Selected Elements:**
- **Original Array:**  
  ```
  [[ 8  2 14  7  6]  → (0th row) ❌ Ignore
   [19 12  3 18  5]  → (1st row) ✅ Include
   [13 11  9  1 17]  → (2nd row) ✅ Include
   [10 20  4  8 16]  → (3rd row) ✅ Include
   [15  6 10 19 14]] → (4th row) ❌ Ignore
  ```
- **Result (`array[1:-1, 0]`):**
  ```
  [19 13 10]
  ```

---

### **Final Output of `border_elements`:**
Yeh **saare border wale elements extract karega**, including **top row, bottom row, left border ke middle elements, aur right border ke middle elements**.

```
[ 8  2 14  7  6  15  6 10 19 14  19 13 10  5 17 16]
```

````python
array = np.random.randint(1,20,size = (4,4))
print("Original Array: ")
print(array)
print("\n")

row_array = np.random.randint(1,20,size=(4,))
print("Row Array:")
print(row_array)
print("\n")

result = array - row_array[:, np.newaxis]
print("Resulting array:")
print(result)
````

Ye code NumPy ka use karke ek 4x4 ka random array generate kar raha hai aur ek 1D row array bhi bana raha hai. Phir in dono arrays ka element-wise subtraction ho raha hai. Chalo step-by-step samajhte hain:

### **Step 1: 4x4 ka random array generate karna**
```python
array = np.random.randint(1,20,size = (4,4))
print("Original Array: ")

```
- Yeh line ek **4x4 ka matrix** (2D array) banati hai jisme har element 1 se 19 ke beech ka ek random integer hota hai.

### **Step 3: Broadcasting ka use karke subtraction**
```python
result = array - row_array[:, np.newaxis]
print("Resulting array:")
print(result)
```
- `row_array[:, np.newaxis]` ka matlab hai ki hum isko **column vector** bana rahe hain (4x1 shape).
- Ab **broadcasting** ki wajah se yeh column vector **4x4 matrix ke har column se subtract ho jayega**.

### **Example Output:**
Maan lo, ye generated arrays hain:
```
Original Array:
[[10 15  6  8]
 [14  7 19  4]
 [ 5 18 11 12]
 [16 13 17  9]]

Row Array:
[3 8 5 2]

Resulting Array:
[[ 7 12  3  5]  # 10-3, 15-3, 6-3, 8-3
 [ 6 -1 11 -4]  # 14-8, 7-8, 19-8, 4-8
 [ 0 13  6  7]  # 5-5, 18-5, 11-5, 12-5
 [14 11 15  7]] # 16-2, 13-2, 17-2, 9-2
```
### **Key Concept:** **Broadcasting**
- NumPy **automatically adjust karta hai shapes** taki operations perform ho sake.
- `row_array[:, np.newaxis]` (4x1) ko **broadcast** karke (4x4) bana diya gaya.
- Phir **har row ke har element se corresponding row_array ka value minus ho gaya**.

Agar broadcasting na hoti, toh **shapes mismatch ka error** aata! 😅

### `row_array[:, np.newaxis]` ka matlab kya hai? 🤔

#### 🔹 **Problem:**
Jab hum `array - row_array` karte hain, toh **shapes mismatch ho sakti hai** kyunki:
- `array` ka shape **(4,4)** hai.
- `row_array` ka shape **(4,)** hai (yeh ek 1D array hai).

Lekin humein **row-wise subtraction** karna hai **column-wise nahi**, isliye humein `row_array` ka **shape change** karna padega.

---

### 🔹 **`[:, np.newaxis]` ka role**
```python
row_array[:, np.newaxis]
```
- `row_array` ka shape originally `(4,)` hota hai, jo **1D array hai**.
- `[:, np.newaxis]` **iska ek extra dimension add kar deta hai**, taaki yeh **(4,1) ka column vector** ban jaaye.
- Yeh NumPy ka ek **trick hai broadcasting ke liye**, taaki subtraction properly ho sake.

---

### 🔹 **Example with Shapes**
```python
import numpy as np

row_array = np.array([3, 8, 5, 2])
print("Original row_array shape:", row_array.shape)

row_array_new = row_array[:, np.newaxis]
print("New shape after np.newaxis:", row_array_new.shape)
```
🔹 **Output:**
```
Original row_array shape: (4,)
New shape after np.newaxis: (4,1)
```
- Pehle `(4,)` tha (1D array).
- `[:, np.newaxis]` lagane se `(4,1)` ban gaya (2D column vector).

---

Ab agar hum `array` se subtract karein:

- Ab NumPy isko **(4,4)** me **broadcast** karega, taki subtraction possible ho.
- Har row ke **har column se corresponding row_array ka element minus ho jayega**.

---

### 🔹 **Summary**
- `row_array[:, np.newaxis]` ka use karke **row ko column me convert** kar rahe hain.
- Isse NumPy **broadcasting enable** karta hai, taki subtraction **har row ke sabhi elements pe ho sake**.
- Agar yeh nahi lagate toh `array - row_array` shape mismatch error de sakta tha.

````python
matrix = np.arange(1,10).reshape(3,3)
print("Matrix")
print(matrix)
print("\n")

determinant = np.linalg.det(matrix)
print("Determinant:", determinant)

inverse = np.linalg.inv(matrix)
print("Inverse:")
print(inverse)

eigenvalues = np.linalg.eigvals(matrix)
print("Eigenvalues:", eigenvalues)
````

Ye code **NumPy ka linear algebra module (`np.linalg`)** use karke ek matrix ke **determinant, inverse, aur eigenvalues** calculate kar raha hai. Chalo step-by-step samajhte hain:

---

### **1️⃣ Determinant nikalna**
```python
determinant = np.linalg.det(matrix)
print("Determinant:", determinant)
```
🔹 **Explanation:**  
- `np.linalg.det(matrix)` kisi bhi **square matrix ka determinant** nikalta hai.
- Determinant ek **single number hota hai** jo batata hai ki matrix **invertible hai ya nahi**.
- Agar `det = 0` ho, to matrix ka **inverse exist nahi karta**.

✅ **Example:**
```python
matrix = np.array([[4, 3], [2, 1]])
determinant = np.linalg.det(matrix)
print("Determinant:", determinant)
```
🔹 **Output:**
```
Determinant: -2.0
```

---

### **2️⃣ Inverse nikalna**
```python
inverse = np.linalg.inv(matrix)
print("Inverse:")
print(inverse)
```
🔹 **Explanation:**  
- `np.linalg.inv(matrix)` kisi bhi **square matrix ka inverse** nikalta hai.
- **Condition:** Matrix **invertible tabhi hoti hai jab uska determinant `0` na ho**.
- Inverse matrix ka matlab hota hai ek aisi matrix jo **original matrix ke saath multiply karne pe identity matrix deti hai**.

✅ **Example:**
```python
matrix = np.array([[4, 3], [2, 1]])
inverse = np.linalg.inv(matrix)
print("Inverse:")
print(inverse)
```
🔹 **Output:**
```
Inverse:
[[-0.5  1.5]
 [ 1.  -2. ]]
```
👉 **Check:** Agar `matrix @ inverse` karenge to **identity matrix milega**.

---

### **3️⃣ Eigenvalues nikalna**
```python
eigenvalues = np.linalg.eigvals(matrix)
print("Eigenvalues:", eigenvalues)
```
🔹 **Explanation:**  
- Eigenvalues (`λ`) **kisi matrix ka special property** hoti hain jo **linear transformations** me kaam aati hain.
- `np.linalg.eigvals(matrix)` **sirf eigenvalues** return karega (eigenvectors nahi).
- Eigenvalues ka formula hota hai:
  \[
  \det(A - \lambda I) = 0
  \]
  Jahan `I` **identity matrix** hai aur `λ` eigenvalues hain.

✅ **Example:**
```python
matrix = np.array([[4, 3], [2, 1]])
eigenvalues = np.linalg.eigvals(matrix)
print("Eigenvalues:", eigenvalues)
```
🔹 **Output:**
```
Eigenvalues: [5.61803399 -0.61803399]
```

---

### **Summary**
| Function | Kya Karta Hai? |
|----------|--------------|
| `np.linalg.det(matrix)` | **Determinant** nikalta hai (0 hoga to inverse nahi milega). |
| `np.linalg.inv(matrix)` | **Inverse matrix** return karta hai (agar determinant **zero nahi** hai to). |
| `np.linalg.eigvals(matrix)` | Matrix ki **eigenvalues** return karta hai. |

Matlab agar **matrix ka determinant 0 ho**, to `np.linalg.inv(matrix)` error dega! 😃

````python
import numpy.ma as ma
# Create a masked array of shape (3, 3) with random integers
array = np.random.randint(1, 21, size=(3, 3))
masked_array = ma.masked_array(array, mask=np.eye(3, dtype=bool))
print("Original array:")
print(array)
print("Masked array:")
print(masked_array)

# Replace the masked elements with the mean of the unmasked elements
mean_unmasked = masked_array.mean()
masked_array = masked_array.filled(mean_unmasked)
print("Modified masked array:")
print(masked_array)
````
