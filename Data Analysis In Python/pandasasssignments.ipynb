## **📌 `pd.merge(df1, df2, on='Key')` Explained! 🚀**  

Yeh Pandas ka **`.merge()` function** use karke **do DataFrames ko join** kar raha hai **common column 'Key' ke basis pe**. Chalo detail me samajhte hain! 🔥  

---

## **🔹 `pd.merge()` ka kaam kya hai?**  
`pd.merge()` Pandas me **SQL JOIN jaisa kaam karta hai**. Yeh **do DataFrames ko ek common column ke basis pe merge karta hai**.  

**Syntax:**  
```python
merged = pd.merge(df1, df2, on='Key')
```
🔹 **`df1`** → Pehla DataFrame  
🔹 **`df2`** → Dusra DataFrame  
🔹 **`on='Key'`** → Dono DataFrames me jo common column 'Key' hai, uske basis pe merge karega  

---

## **🔹 Example: Merge Two DataFrames**  
```python
import pandas as pd

# First DataFrame
df1 = pd.DataFrame({
    'Key': ['A', 'B', 'C'],
    'Value1': [100, 200, 300]
})

# Second DataFrame
df2 = pd.DataFrame({
    'Key': ['A', 'B', 'D'],
    'Value2': [10, 20, 40]
})

# Merging both DataFrames on 'Key'
merged = pd.merge(df1, df2, on='Key')

print(merged)
```

👉 **Output:**
```
  Key  Value1  Value2
0   A     100      10
1   B     200      20
```
✅ **Sirf woh rows aayi jisme 'Key' common tha (A & B).**  
✅ **'C' (df1 me tha) aur 'D' (df2 me tha) nahi aaye kyunki yeh default `inner join` hota hai.**  

---

## **🔹 Different Merge Types**
Agar **sare records lana hai** toh **different types of joins** use kar sakte hain:  

### **1️⃣ `how='inner'` (Default) → Sirf Common Values**
```python
merged = pd.merge(df1, df2, on='Key', how='inner')
```
✅ **Jo values dono me common hain, sirf wahi aayengi.** (Default behavior)  

---

### **2️⃣ `how='outer'` → Sabhi Rows Aayengi (`NaN` values ke saath)**
```python
merged = pd.merge(df1, df2, on='Key', how='outer')
```
👉 **Output:**
```
  Key  Value1  Value2
0   A   100.0    10.0
1   B   200.0    20.0
2   C   300.0     NaN
3   D     NaN    40.0
```
✅ **Jo values kisi ek DataFrame me bhi hain, sabhi aayengi.**  
✅ **Missing values ko `NaN` se fill karega.**  

---

### **3️⃣ `how='left'` → Left Table ke Sare Rows**
```python
merged = pd.merge(df1, df2, on='Key', how='left')
```
👉 **Output:**
```
  Key  Value1  Value2
0   A     100    10.0
1   B     200    20.0
2   C     300     NaN
```
✅ **df1 ka pura data aayega, df2 se sirf matching rows.**  

---

### **4️⃣ `how='right'` → Right Table ke Sare Rows**
```python
merged = pd.merge(df1, df2, on='Key', how='right')
```
👉 **Output:**
```
  Key  Value1  Value2
0   A   100.0      10
1   B   200.0      20
2   D     NaN      40
```
✅ **df2 ka pura data aayega, df1 se sirf matching rows.**  

---

## **📌 Summary Table**
| Merge Type | Description |
|------------|------------|
| **Inner Join** (`how='inner'`) | **Sirf Common Values** |
| **Outer Join** (`how='outer'`) | **Sare Rows (Missing ko NaN)** |
| **Left Join** (`how='left'`) | **df1 ka pura data, df2 se sirf matching** |
| **Right Join** (`how='right'`) | **df2 ka pura data, df1 se sirf matching** |

---

## **🔥 Final Takeaway**
✔ **`pd.merge(df1, df2, on='Key')` inner join karta hai by default.**  
✔ **Agar sare records chahiye toh `how='outer'`, `how='left'`, ya `how='right'` use karo.**  
✔ **SQL `JOIN` ki tarah kaam karta hai!**
