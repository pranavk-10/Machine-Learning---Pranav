## **🔥 Flask me GET aur POST Method Kya Hote Hain?**  

Flask me **GET aur POST** methods ka use hota hai **client (browser) aur server ke beech data exchange karne ke liye**.  

- **GET** → Server se data **mangta** hai (Read-only, URL me dikhta hai).  
- **POST** → Server ko **data bhejta** hai (Secure hota hai, URL me nahi dikhta).  

---

## **⚡ GET vs POST (Ek Quick Comparison)**  

| **Feature**       | **GET** | **POST** |
|------------------|--------|---------|
| **Use Case** | Data fetch karna | Data submit/update karna |
| **Data dikhta hai?** | Haan, URL me dikhta hai | Nahi, hidden hota hai |
| **Security** | Kam secure (data URL me hota hai) | Zyada secure |
| **Size Limit** | Limited (URL me fit hone jitna) | Zyada data bhej sakta hai |
| **Caching** | Yes, browser cache kar sakta hai | No, cache nahi hota |

---

## **🚀 Flask GET Method Example**  
```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/search", methods=["GET"])  
def search():
    query = request.args.get("q")  # URL se query parameter uthana
    if query:
        return f"Search results for: {query}"
    return "Please provide a search query."

if __name__ == "__main__":
    app.run(debug=True)
```

🔹 **Agar tu browser me yeh URL open karega:**  
```
http://127.0.0.1:5000/search?q=flask
```
🔹 **Output aayega:**  
```
Search results for: flask
```
👉 **GET method URL me parameters bhejta hai (`?q=flask`) jo sabko dikhte hain.**

---

## **🚀 Flask POST Method Example**
POST method **form ya API se data bhejne ke liye hota hai**.

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/submit", methods=["POST"])
def submit():
    name = request.form.get("name")  # Form se naam uthana
    return f"Data received: {name}"

if __name__ == "__main__":
    app.run(debug=True)
```

🔹 **Isko test karne ke liye tu `curl` ya Postman use kar sakta hai:**  
```bash
curl -X POST -d "name=Pannu" http://127.0.0.1:5000/submit
```
🔹 **Output:**  
```
Data received: Pannu
```
👉 **POST method ka data URL me nahi dikhta, sirf request body me bheja jata hai.**

---

## **🔥 GET + POST Ek Saath**
Agar tu **GET aur POST dono** ek hi function me allow karna chahta hai:

```python
from flask import Flask, request

app = Flask(__name__)

@app.route("/login", methods=["GET", "POST"])
def login():
    if request.method == "POST":
        username = request.form.get("username")
        password = request.form.get("password")
        return f"Login successful for {username}!"
    return "Please submit the login form."

if __name__ == "__main__":
    app.run(debug=True)
```
### **🔹 Kaise Kaam Karega?**
- **GET Request:** `http://127.0.0.1:5000/login` open karne pe **login form message milega**.  
- **POST Request:**  
  ```bash
  curl -X POST -d "username=pannu&password=1234" http://127.0.0.1:5000/login
  ```
  **Output:**  
  ```
  Login successful for pannu!
  ```

✔ **GET → Server se data mangaane ke liye (URL me dikhta hai, less secure)**  
✔ **POST → Server ko data bhejne ke liye (Secure hota hai, URL me nahi dikhta)**  
✔ **`request.args.get()` → GET method me query parameters lene ke liye**  
✔ **`request.form.get()` → POST method me form data lene ke liye**

### **🚀 Step 4: Flask App Code likh (AI Response Generate Karne Ke Liye)**
```python
import os
from flask import Flask, request, jsonify
from dotenv import load_dotenv
import openai

# API Key ko load kar raha hai .env file se
load_dotenv()
openai.api_key = os.getenv("OPENAI_API_KEY")

app = Flask(__name__)

@app.route("/chat", methods=["POST"])
def chat():
    try:
        data = request.get_json()
        user_input = data.get("message")

        # OpenAI se response lena
        response = openai.ChatCompletion.create(
            model="gpt-3.5-turbo",  
            messages=[{"role": "user", "content": user_input}]
        )

        reply = response["choices"][0]["message"]["content"]
        return jsonify({"reply": reply})

    except Exception as e:
        return jsonify({"error": str(e)})

if __name__ == "__main__":
    app.run(debug=True)
```

## **1️⃣ Flask Import & Initialization**
```python
from flask import Flask, render_template, request

app = Flask(__name__)
```
- **`Flask`** → Flask framework import kiya.  
- **`render_template`** → HTML templates (index.html, about.html, form.html) render karne ke liye.  
- **`request`** → Form se user input lene ke liye.  
- **`app = Flask(__name__)`** → Flask app initialize kiya.  

---

## **2️⃣ Homepage Route (`/`)**
```python
@app.route('/')
def welcome():
    return '<html><h1>WELCOME TO PANNUS SITE</h1></html>'
```
- **Route** → Jab user `/` (homepage) pe aayega toh `WELCOME TO PANNUS SITE` show hoga.  
- **HTML inside Python** → Directly HTML return kiya hai, but best practice **template use karna** hai.  

---

## **3️⃣ Index Page (`/index`)**
```python
@app.route('/index', methods=['GET'])
def index():
    return render_template('index.html')
```
- **Route** → `/index` pe jaane par `index.html` render hoga.  
- **`methods=['GET']`** → Sirf GET request accept karega.  

---

## **4️⃣ About Page (`/about`)**
```python
@app.route('/about')
def about():
    return render_template('about.html')
```
- **Route** → `/about` pe jaane par `about.html` render hoga.  
- **GET method** → Default hota hai, isliye explicitly likhne ki zaroorat nahi hai.  

---

## **5️⃣ Form Handling (`/form`)**
```python
@app.route('/form', methods=['GET', 'POST'])
def form():
    if request.method == 'POST':
        name = request.form['name']
        return f"HELLO {name}!"
    return render_template('form.html')
```
- **Route** → `/form` page handle karega.  
- **GET Request** → `form.html` render hoga (jab page khulega).  
- **POST Request** → Jab user **form submit karega**, toh:
  - `name = request.form['name']` → Form ka **name** value extract karega.
  - `return f"HELLO {name}!"` → User ko greeting message return karega.  

---

## **6️⃣ Running The Flask App**
```python
if __name__ == '__main__':
    app.run(debug=True)
```
- **`if __name__ == '__main__'`** → Ensure karta hai ki script **direct run ho** (`import` na ho).  
- **`debug=True`** → Debug mode enable karega, toh **errors show honge & auto-restart hoga** jab code change karega.

   - `/about` → `about.html` render hoga.
   - `/form` → Ek form dikhega, naam dalke submit karo → "HELLO [Name]!"
