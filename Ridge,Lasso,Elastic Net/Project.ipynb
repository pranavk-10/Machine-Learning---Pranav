```python
df_copy = pd.read_csv('cleaned_data.csv')
```

```python
df_copy.head()
```

```python
df_copy = df_copy.drop(['day','month','year'],axis=1)
```

```python
df_copy.head()
```

```python
df_copy['Classes']=np.where(df_copy['Classes'].str.contains('not fire'),0,1)
```

```python
df_copy.head()
```

```python
df_copy['Classes'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.style.use('seaborn-v0_8')  # safer for latest versions (Matplotlib 3.6+)
df_copy.hist(bins=50, figsize=(20, 15))
plt.tight_layout()
plt.show()
```

```python
percentage = df_copy['Classes'].value_counts(normalize=True)*100
```

```python


# Sample data
labels = ['Fire', 'Not Fire']

# Plotting pie chart
plt.figure(figsize=(12,6))
plt.pie(percentage, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff6b6b','#4ecdc4'])
plt.title('Distribution of Fire vs Not Fire')
#plt.axis('equal')  # Equal aspect ratio makes it a circle
plt.show()
```
