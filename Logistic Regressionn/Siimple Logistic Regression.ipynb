## Simple Logistic Regression


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import make_classification
```

```python
x,y=make_classification(n_samples=1000,n_features=10,n_classes=2,random_state=42) #creating datasets
```

```python
x
```

```python
pd.DataFrame(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
reg = LogisticRegression()
```

```python
reg.fit(x_train,y_train)
```

```python
y_pred_test = reg.predict(x_test)
```

```python
print(y_pred_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

```python
score = accuracy_score(y_test,y_pred_test)
print(score)
cm = confusion_matrix(y_test,y_pred_test)
print(cm)
report = classification_report(y_test,y_pred_test)
print(report)
```

### Hyperparameter tunning


```python
model = LogisticRegression()
```

```python
penalty = ['l1','l2','elasticnet']
c_values = [0.1, 0.5, 1.0, 5.0]
solver = ['newton-cg','lbfgs','liblinear','sag','saga']
```

```python
parameters = dict (penalty = penalty, C= c_values , solver = solver)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import StratifiedKFold
cv = StratifiedKFold()
```

```python
grid = GridSearchCV(estimator=model,param_grid=parameters,scoring='accuracy',cv=cv,n_jobs=-1)
```

```python
grid
```

```python
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
y_new_pred = grid.predict(x_test)
```

```python
y_new_pred
```

```python
score = accuracy_score(y_test,y_new_pred)
print(score)
cm = confusion_matrix(y_test,y_new_pred)
print(cm)
report = classification_report(y_test,y_new_pred)
print(report)
```

### Randomised CV

```python
from sklearn.model_selection import RandomizedSearchCV
random = RandomizedSearchCV(estimator= model,param_distributions=parameters,cv=5,scoring='accuracy')
```

```python
random.fit(x_train,y_train)
```

```python
random.best_params_
```

```python
random.best_score_
```

```python
y_rand_pred = random.predict(x_test)
```

```python
score = accuracy_score(y_test,y_rand_pred)
print(score)
cm = confusion_matrix(y_test,y_rand_pred)
print(cm)
report = classification_report(y_test,y_rand_pred)
print(report)
```
