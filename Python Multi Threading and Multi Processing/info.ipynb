## **Process, Thread aur Program ka Difference**

### **ğŸ”¥ Example Code (Multiprocessing using Multiple Cores)**  
```python
import multiprocessing
import time

def task1():
    print("Task 1 running on:", multiprocessing.current_process().name)
    time.sleep(2)
    print("Task 1 Completed")

def task2():
    print("Task 2 running on:", multiprocessing.current_process().name)
    time.sleep(2)
    print("Task 2 Completed")

if __name__ == "__main__":
    p1 = multiprocessing.Process(target=task1, name="Core 1 Process")
    p2 = multiprocessing.Process(target=task2, name="Core 2 Process")

    p1.start()
    p2.start()

    p1.join()
    p2.join()

    print("Both tasks completed")
```

1. **Imports**: Sabse pehle, `multiprocessing`, `time`, `math` aur `sys` modules ko import kiya gaya hai.
   - `multiprocessing`: Parallel processes ko manage karne ke liye.
   - `time`: Execution time ko measure karne ke liye.
   - `math`: Factorial calculation ke liye.
   - `sys`: Maximum integer digits ko manage karne ke liye.

2. **factorial Function**:
   - Yeh function ek number ko accept karta hai aur uska factorial calculate karta hai.
   - **sys.set_int_max_str_digits(10000)** ka use yeh ensure karne ke liye ki hum bahut bade integers handle kar sakein bina kisi overflow ke.

3. **Factorial Calculation**: `math.factorial(n)` ka use factorial compute karne ke liye kiya gaya hai.

```python
if __name__ == "__main__":
    # List of numbers for which factorial will be calculated
    numbers = [50, 60, 700, 300]  
    start_time = time.time()  # Start time to measure execution

    # Creating a pool of processes to run factorial function in parallel
    with multiprocessing.Pool() as pool:
        results = pool.map(factorial, numbers)  # Map the function to the list of numbers

    end_time = time.time()  # End time to measure execution

    # Print the results and total time taken
    print(f"Results : {results}")
    print(f"Total Time Taken : {end_time - start_time} seconds")
```

4. **Multiprocessing Pool**:
   - **Pool** ek group of processes hai jo parallel execution ko manage karta hai.
   - `multiprocessing.Pool()` ek pool create karta hai jo multiple processes ko handle karta hai.
   - `pool.map(factorial, numbers)` ka use factorial function ko parallel execute karne ke liye kiya gaya hai. Yeh function list of numbers ko input deta hai aur har number ke liye factorial calculate karta hai.

5. **Execution Time**: 
   - `start_time` aur `end_time` ke beech ka difference calculate karke total execution time ko print kiya gaya hai.

6. **Output**:
   - Jab process complete hota hai, toh `results` list mein sabhi factorials store hote hain, jo print kiye jaate hain.
   - Total time jo process ko complete karne mein lagta hai, woh bhi print hota hai.

### Key Concepts:
- **Multiprocessing Pool**: Yeh processes ko manage karta hai aur unko efficiently parallel execute karta hai.
- **CPU-bound Tasks**: Jahan CPU intensive operations hoti hain (jaise factorial calculation), multiprocessing ka use best hota hai.
- **Parallel Execution**: Multiple processes ko ek saath run karna, jisse task jaldi complete hota hai.
- **Efficiency**: Multiprocessing mein har process apne independent memory space ka use karta hai, isliye **Global Interpreter Lock (GIL)** ka issue nahi hota, jo threading mein hota hai.

### Benefits of Using Multiprocessing:
1. **Independent Processes**: Har process apne memory space mein kaam karta hai, isliye synchronization issues kam hote hain.
2. **Faster Execution**: CPU-bound tasks ko parallel mein run karna zyada fast hota hai as compared to sequential execution.
3. **No GIL Limitations**: Python mein **GIL** threading mein problem create karta hai jab multiple threads ek saath CPU-bound task perform karte hain, lekin multiprocessing mein yeh issue nahi hota.
