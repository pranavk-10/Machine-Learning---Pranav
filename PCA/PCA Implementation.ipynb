### PCA Implementation

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
print(cancer['DESCR'])
```

```python
print(cancer.keys())
```

```python
print(cancer['data'])
print(cancer['feature_names'])
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scalar = StandardScaler()
```

```python
scalar.fit(df)
```

```python
scaled_data = scalar.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
data_pca=pca.fit_transform(scaled_data)
```

```python
data_pca
```

```python
pd.DataFrame(data_pca)
```

```python
pca.explained_variance_
```

```python
plt.figure(figsize=(10,8))
plt.scatter(data_pca[:,0], data_pca[:,1], c=cancer['target'], cmap='plasma')
plt.xlabel("First Principal Component")
plt.ylabel("Second Principal Component")
plt.title("PCA - Breast Cancer Dataset")
plt.colorbar(label='Target (0 = Malignant, 1 = Benign)')

# Optional: annotate the first point
plt.annotate("Sample 0", (data_pca[0,0], data_pca[0,1]), color='green', fontsize=12)
plt.show()
```
