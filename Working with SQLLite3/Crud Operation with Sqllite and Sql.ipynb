---

### **1Ô∏è‚É£ `cursor.execute(''' ... SQL QUERY ... ''')`**  
‚û°Ô∏è **Matlab:** `cursor.execute()` ka use karke ek **SQL query run** kar rahe hain.   
‚û°Ô∏è **Kya ho raha hai?**  
   - `CREATE TABLE IF NOT EXISTS Employees` ‚Üí Ek **table banayenge**, agar pehle se nahi hai toh.

---

### **2Ô∏è‚É£ `CREATE TABLE IF NOT EXISTS Employees (`**  
‚û°Ô∏è **Matlab:** Agar `"Employees"` naam ka table **nahi hai**, toh **naya banayega**.  
‚û°Ô∏è **Simple shabdon me:**  
   - Jaise **naya folder tabhi banta hai jab pehle se exist nahi karta**,  
   - Waise hi `IF NOT EXISTS` ka matlab hai ki **agar table pehle se hai, toh dobara mat banao**.  

---

### **3Ô∏è‚É£ Table ke andar Columns define ho rahe hain:**  
Yeh **4 columns** ban rahe hain:  

#### **üîπ `id INTEGER PRIMARY KEY,`**  
‚úÖ **Matlab:** `id` ek **unique identifier** hoga, jo **automatic** alag-alag har row ke liye set hoga.  
‚úÖ **Kyon?** Kyunki **`PRIMARY KEY`** ka rule hai ki:  
   - **Har row ka `id` alag hoga (unique)**  
   - **Khali nahi ho sakta (null nahi ho sakta)**  
   - **Auto-increment hota hai agar explicitly insert na karein**  

‚û°Ô∏è **Example:**  
| id | name  | age | occupation |
|----|------|----|------------|
| 1  | Aman | 25 | Engineer   |
| 2  | Ravi | 30 | Doctor     |

---

#### **üîπ `name TEXT NOT NULL,`**  
‚úÖ **Matlab:** `name` ek **TEXT** (string) hoga aur **empty nahi ho sakta**.  
‚úÖ **Kyon?** `"NOT NULL"` ka matlab hai ki **har entry me `name` zaroor hona chahiye**.  

‚û°Ô∏è **Example (‚úÖ Valid Data):**  
| id | name  | age | occupation |
|----|------|----|------------|
| 1  | Aman | 25 | Engineer   |

‚û°Ô∏è **Example (‚ùå Invalid Data - Error aayega!):**  
| id | name  | age | occupation |
|----|------|----|------------|
| 2  | (empty) | 30 | Doctor |  ‚ùå **Error! `name` cannot be NULL**  

---

#### **üîπ `age INTEGER,`**  
‚úÖ **Matlab:** `age` ek **integer (number)** hoga.  
‚úÖ **Koi restriction nahi hai**, empty ho sakta hai (NULL allowed hai).  

‚û°Ô∏è **Example:**  
| id | name  | age | occupation |
|----|------|----|------------|
| 1  | Aman | 25 | Engineer   |
| 2  | Ravi | NULL | Doctor  |

---

#### **üîπ `occupation TEXT`**  
‚úÖ **Matlab:** `occupation` ek **TEXT (string)** hoga.  
‚úÖ **Ye bhi NULL ho sakta hai**, yaani agar kisi ka occupation nahi pata, toh **blank** chhod sakte hain.  

‚û°Ô∏è **Example:**  
| id | name  | age | occupation |
|----|------|----|------------|
| 1  | Aman | 25 | Engineer   |
| 2  | Ravi | 30 | (NULL) |

---

### **4Ô∏è‚É£ `connection.commit()`**  
‚úÖ **Matlab:** Jo bhi **changes database me kiye gaye hain, unko save kar raha hai**.  
‚úÖ **Kyon?**  
   - SQLite ek **transaction-based system** hai.  
   - Agar aap `commit()` nahi likhoge, toh **table ya data permanent save nahi hoga**.  
   - **Jaise Microsoft Word me changes save karne ke liye "Ctrl + S" karte hain, waise hi `commit()` database me changes save karta hai.**  

---

### ‚úÖ **Final Summary (Hinglish me)**  
1Ô∏è‚É£ **Ek table banane ki SQL query likhi gayi hai.**  
2Ô∏è‚É£ Agar `"Employees"` table **pehle se exist nahi karta**, toh **naya table create hoga**.  
3Ô∏è‚É£ **4 columns banaye gaye hain:**  
   - `id INTEGER PRIMARY KEY` ‚Üí Unique ID  
   - `name TEXT NOT NULL` ‚Üí Name **empty nahi ho sakta**  
   - `age INTEGER` ‚Üí Age, number hoga, empty ho sakta hai  
   - `occupation TEXT` ‚Üí Job title, empty ho sakta hai  
4Ô∏è‚É£ **Database me changes save karne ke liye `commit()` lagaya gaya hai.**

Aapke **SQL queries** me **multiple syntax errors** hain. Chaliye ek-ek error ko **identify** karte hain aur **corrected version** bhi dete hain.  

---

## **üî¥ Error 1: Extra comma after `Employees(name, age, occupation),`**
```sql
Insert into Employees(name,age,occupation),
             values("Pranav",19,"Engineer)
```
### **‚ùå Error:**
- `Insert into Employees(name,age,occupation),` ‚Üí Yaha **extra comma `,` galat hai**.  
- **Correct syntax:** `INSERT INTO Employees (columns) VALUES (values);`

### ‚úÖ **Corrected Version**
```sql
cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Pranav", 19, "Engineer")
''')
```
---

## **üî¥ Error 2: Same issue in second query (extra comma)**
```sql
Insert into Employees(name,age,occupation),
             values("Tanishka",19,"Engineer)
```
### ‚úÖ **Corrected Version**
```sql
cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Tanishka", 19, "Engineer")
''')
```
---

## **üî¥ Error 3: Wrong separator `;` instead of `,`**
```sql
Insert into Employees(name;age;occupation),
             values("Ojas",19,"Engineer)
```
### **‚ùå Error:**
- Column names **`name;age;occupation`** ‚Üí **Yaha `;` nahi hona chahiye, `,` hona chahiye.**

### ‚úÖ **Corrected Version**
```sql
cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Ojas", 19, "Engineer")
''')
```
---

## **üî¥ Error 4: Extra `"` in `18'"` and missing closing quote for `"CA`**
```sql
Insert into Employees(name,age,occupation),
             values("Kirti",18',"CA)
```
### **‚ùå Errors:**
1. `18'",` ‚Üí **Yaha `"` galat jagah par hai.**  
2. `"CA` ‚Üí **Closing `"` missing hai.**

### ‚úÖ **Corrected Version**
```sql
cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Kirti", 18, "CA")
''')
```
---

## **üîµ Final Corrected Code (All Queries Fixed)**
```python
cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Pranav", 19, "Engineer")
''')

cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Tanishka", 19, "Engineer")
''')

cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Ojas", 19, "Engineer")
''')

cursor.execute('''
INSERT INTO Employees (name, age, occupation)
VALUES ("Kirti", 18, "CA")
''')
```
---

## **‚úÖ Summary (Kya Galat Tha?)**
1. **Extra comma (`,`)** ‚Üí `Employees(name,age,occupation),` **‚ùå Wrong**  
   **‚úÖ Correct:** `Employees(name, age, occupation)`
2. **Wrong separator (`;` instead of `,`)** ‚Üí `name;age;occupation` **‚ùå Wrong**  
   **‚úÖ Correct:** `name, age, occupation`
3. **Quotes issues (`"` misplaced/missing)** ‚Üí `18'",` **‚ùå Wrong**  
   **‚úÖ Correct:** `18, "CA"`

````python
cursor.execute('Select * from Employees')
rows = cursor.fetchall()

for row in rows:
    print(row)
````

Yeh code SQLite database se **Employees** table ka sara data retrieve karne aur print karne ke liye hai. Chaliye **line-by-line** iska explanation dekhte hain.  

---

### **üìù Code Breakdown:**
```python
cursor.execute('SELECT * FROM Employees')
```
- `cursor.execute()` ‚Üí **SQL command run karne ke liye hota hai.**
- `'SELECT * FROM Employees'` ‚Üí  
  - **`SELECT *`** ‚Üí Table ka **sara data** fetch karega.  
  - **`FROM Employees`** ‚Üí **Employees table** se data uthayega.

---

```python
rows = cursor.fetchall()
```
- **`cursor.fetchall()`** ‚Üí  
  - `SELECT` query ka **saara result ek list of tuples** me store karega.
  - **Example output:**  
    ```python
    [
        (1, "Pranav", 19, "Engineer"),
        (2, "Tanishka", 19, "Engineer"),
        (3, "Ojas", 19, "Engineer"),
        (4, "Kirti", 18, "CA")
    ]
    ```
  - Har row **ek tuple** ke form me aayegi.

---

```python
for row in rows:
    print(row)
```
- **`for row in rows:`** ‚Üí Har row ko ek-ek karke iterate karega.
- **`print(row)`** ‚Üí **Row ko print karega** (jo ek tuple hoga).
